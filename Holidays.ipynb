# Loading Booking data

```python
db_data = pd.read_csv("OPENDATA_BOOKING_CARSHARING.csv", sep = ';', header = 0)
```

```python
db_data.head()
```

# Considering the data we need

```python
data = ['COMPUTE_EXTRA_BOOKING_FEE','TRAVERSE_USE','CATEGORY_HAL_ID','DISTANCE','START_RENTAL_ZONE','START_RENTAL_ZONE_HAL_ID','END_RENTAL_ZONE','RENTAL_ZONE_HAL_SRC','END_RENTAL_ZONE_HAL_ID','TECHNICAL_INCOME_CHANNEL']
db_data.drop(data,inplace = True, axis = 1)
```

```python
db_data = db_data.rename(columns = {'CITY_RENTAL_ZONE' :'CITY'})
db_data = db_data.rename(columns = {'VEHICLE_HAL_ID' :'VEHICLE_TYPE_ID'})
db_data = db_data.rename(columns = {'BOOKING_HAL_ID' :'BOOKING_ID'})
db_data.head()
```

# Seperating Dates from the colmns

```python
# DATE BOOKING column
db_data['DATE_BOOKING'] = pd.to_datetime(db_data.DATE_BOOKING)
db_data['bookingDate'] = db_data.DATE_BOOKING.dt.date
db_data['bookingDate'] = pd.to_datetime(db_data.bookingDate)
db_data['BWD'] =db_data.DATE_BOOKING.dt.weekday #where 0-->SUNDAY until 6-->SATURDAY
# DATE_FROM column
db_data['DATE_FROM'] = pd.to_datetime(db_data.DATE_FROM)
db_data['fromDate'] = db_data.DATE_FROM.dt.date
db_data['fromDate'] = pd.to_datetime(db_data.fromDate)
db_data['FWD'] =db_data.DATE_FROM.dt.weekday

# DATE_UNTIL
db_data['DATE_UNTIL'] = pd.to_datetime(db_data.DATE_UNTIL)
db_data['untilDate'] = db_data.DATE_UNTIL.dt.date
db_data['untilDate'] = pd.to_datetime(db_data.untilDate)
db_data['UWD'] =db_data.DATE_UNTIL.dt.weekday

db_data
```

```python
new = ['DATE_BOOKING','DATE_FROM','DATE_UNTIL']
db_data.drop(new,inplace = True, axis = 1)
db_data
```
