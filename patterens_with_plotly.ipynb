```python
import plotly.offline as py
import plotly.graph_objs as go

py.init_notebook_mode(connected=True)

# Create traces
trace0 = go.Scatter(
    x = per_day.BOOKING_DATE,
    y = per_day.BOOKING_HAL_ID,
    mode = 'lines+markers',
    name = 'number of bookings'
)
trace1 = go.Scatter(
    x = per_day.BOOKING_DATE,
    y = per_day.CUSTOMER_HAL_ID,
    mode = 'lines+markers',
    name = 'number of customers'
)
trace2 = go.Scatter(
    x = per_day.BOOKING_DATE,
    y = per_day.VEHICLE_HAL_ID,
    mode = 'lines',
    name = 'number of cars'
)

fig = go.Figure(data=[trace0, trace1, trace2])
py.iplot(fig, filename="overall timeseries plot")
```

```python
simple_per_state.reset_index(inplace=True)
```

```python
simple_per_state.head()
```

```python
#plotting for each state
trace_bookings = go.Bar(x=simple_per_state.FEDERAL_STATE,
                  y=simple_per_state.BOOKING_HAL_ID,
                  name='bookings',
                  marker=dict(color='#ffcdd2'))

trace_customers = go.Bar(x=simple_per_state.FEDERAL_STATE,
                y=simple_per_state.CUSTOMER_HAL_ID,
                name='customers',
                marker=dict(color='#A2D5F2'))

trace_vehicles = go.Bar(x=simple_per_state.FEDERAL_STATE,
                y=simple_per_state.VEHICLE_HAL_ID,
                name='cars',
                marker=dict(color='#59606D'))
trace_zones = go.Bar(x=simple_per_state.FEDERAL_STATE,
                y=simple_per_state.START_RENTAL_ZONE,
                name='zones',
                marker=dict(color='#00CED1'))
layout = go.Layout(title="stats per state",
                xaxis=dict(title='State')
                )
fig = go.Figure(data=[trace_bookings, trace_customers, trace_zones, trace_vehicles], layout=layout)
py.iplot(fig, filename="overall perstate plot")
```
